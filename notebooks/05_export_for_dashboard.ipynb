```python
# ============================================================
# NOTEBOOK 05 — Export for Dashboard
# ============================================================
# Reads all Gold and ML Parquet tables and exports them as
# single CSV files to a dedicated export volume path.
# These CSVs will be downloaded and committed to GitHub
# for the Streamlit dashboard to read.
# ============================================================

from pyspark.sql import functions as F

GOLD_PATH = "/Volumes/workspace/default/flight_raw_data/gold"
ML_PATH = "/Volumes/workspace/default/flight_raw_data/ml"
EXPORT_PATH = "/Volumes/workspace/default/flight_raw_data/exports"

print("Gold input path:", GOLD_PATH)
print("ML input path:", ML_PATH)
print("Export output path:", EXPORT_PATH)
```

```python
# ============================================================
# Step 1 — Read all Gold and ML Parquet tables
# ============================================================

gold_airline = spark.read.parquet(f"{GOLD_PATH}/airline_delays")
gold_routes = spark.read.parquet(f"{GOLD_PATH}/route_delays")
gold_monthly = spark.read.parquet(f"{GOLD_PATH}/monthly_trend")
gold_hourly = spark.read.parquet(f"{GOLD_PATH}/hourly_delays")
ml_metrics = spark.read.parquet(f"{ML_PATH}/metrics")
ml_importance = spark.read.parquet(f"{ML_PATH}/feature_importance")

print("Gold airline rows:", gold_airline.count())
print("Gold routes rows:", gold_routes.count())
print("Gold monthly rows:", gold_monthly.count())
print("Gold hourly rows:", gold_hourly.count())
print("ML metrics rows:", ml_metrics.count())
print("ML importance rows:", ml_importance.count())
```

```python
# ============================================================
# Step 2 — Export all tables as single CSV files
# ============================================================
# coalesce(1) merges all Spark partitions into one single file.
# Without it, Spark writes multiple part-00000, part-00001
# files which are hard to download and use in Streamlit.
# header=True ensures column names are written in the CSV.
# ============================================================

print("Exporting airline delays...")
gold_airline.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/airline_delays")

print("Exporting route delays...")
gold_routes.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/route_delays")

print("Exporting monthly trend...")
gold_monthly.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/monthly_trend")

print("Exporting hourly delays...")
gold_hourly.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/hourly_delays")

print("Exporting ML metrics...")
ml_metrics.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/ml_metrics")

print("Exporting ML feature importance...")
ml_importance.coalesce(1).write.mode("overwrite").option("header", True).csv(f"{EXPORT_PATH}/ml_importance")

print("All exports complete.")
```
